## 1) Nivel socioeconómico

El nivel socioeconómico en Chile tiene siete clases: AB (clase alta), C1a (clase media acomodada), C1b (clase media emergente), C2 (clase media típica), C3 (clase media baja), D (clase media vulnerable) y E (pobres).

Elegir un algoritmo de Clustering y formar 7 grupos usando como datos las columnas `'longitud', 'latitud', 'nse'`.

Entregar el gráfico anterior del mapa geográfico coloreando cada coordenada según etiqueta del Clustering.

Analizar los resultados obtenidos. ¿Es posible identificar en qué sector de Santiago se concentra la riqueza?

```python
from sklearn.cluster import KMeans
x = df_barrio[["longitud","latitud","nse"]] ##Datos Pedidos
df_kmeans=x.copy() #Se crea una copia
clustering = KMeans(n_clusters=7,max_iter=300) #Se aplica Kmeans con 7 cluster y 300 Iteraciones
clustering.fit(df_kmeans)

```

```python
df_kmeans['kmeans_cluster'] = clustering.labels_
df_kmeans
```

```python
ab = df_stgo.plot(figsize=(18,16), color='#EFEFEF', edgecolor='#444444')
plt.scatter(data=df_kmeans,x="longitud",y="latitud",c="kmeans_cluster")
plt.title('Kmeans')
plt.xlabel('longitud')
```

- Con el entrenamiento hecho podemos observar claramente los sectores en santiago donde es posible que se concentre la mayor cantidad de riquezas o NSE debido a una zona en la parte superior derecha donde solo existe un tipo de color o cluster.


## 2) Plebiscito

El Plebiscito tenía las opciones Apruebo o Rechazo.

Elegir un algoritmo de Clustering diferente al Punto 1 y formar 2 grupos usando como datos las columnas `'longitud', 'latitud', 'prop_apruebo_2020'`.

Entregar el gráfico anterior del mapa geográfico coloreando cada coordenada según etiqueta del Clustering.

Analizar los resultados obtenidos. ¿En qué sectores de Santiago se "Aprueba" y en qué sectores se "Rechaza"?

```python
from sklearn.cluster import DBSCAN
y = df_barrio[["longitud","latitud","prop_apruebo_2020"]] #Datos Pedidos
df_dbscan=y.copy() #Copia
clustering_dbscan=DBSCAN(eps=0.1,min_samples=450).fit_predict(df_dbscan) #Se aplica DBSCAN con radio de 0.1 y 450 muestras por vecindario
```

```python
ac = df_stgo.plot(figsize=(18,16), color='#EFEFEF', edgecolor='#444444')
colors=["Yellow","Purple"]
labels=["Apruebo","Rechazo"]
plt.scatter(data=df_dbscan,x="longitud",y="latitud",c=clustering_dbscan,s=100)
plt.legend([plt.scatter([],[], color=c, label=l) for c, l in zip(colors, labels)], labels, loc="upper left")    
plt.title('BDSCAN')
plt.xlabel('longitud')
```

- Relacionando los datos entre el problema 1 y el problema 2 podemos decir que las personas que rechazan o donde se concentra la mayor cantidad de personas que rechazan coinciden con la misma zona que tiene mayor nivel socioeconomico
