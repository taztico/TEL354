# And the Oscar Goes To...

**Objetivo**

El objetivo del Taller 2 es predecir la película ganadora del Oscar 2021, entrenando un modelo de regresión logística que use como datos de entrenamiento todas las películas que participaron en la ceremonia antes de ese año.

1. Cargar los datos del archivo `oscar.csv` a un DataFrame y mostrarlo.

```python
import pandas as pd
```

```python
df_oscar=pd.read_csv('oscar.csv')
print("Datos oscar.csv")
df_oscar
```

2. Generar un conjunto de entrenamiento `X_train`, `y_train`:
- La variable objetivo `y_train` será la columna `Winner`, ya que, queremos predecir si una película ganará o perderá el premio.
- La variable `X_train` será todas las demás columnas del DataFrame con excepción del nombre de la película (columna `Film`) porque en realidad el título no debería influir en el resultado de la votación y además habría que codificarlo para usarlo como característica.

```python
X_train=df_oscar.iloc[:,2:]
X_train
```

```python
Y_train=df_oscar['Winner']
Y_train
```

3. Realizar una LogisticRegression usando los datos de entrenamiento `X_train`, `y_train`.

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

4. Generar un conjunto de testing `X_test` con todas las columnas del archivo `nominees2021.csv`, excepto la primera columna que contiene los nombres de las películas. Estas fueron las películas nominadas al premio Oscar el año 2021 y no están en el conjunto de entrenamiento. Por lo tanto, se usarán para realizar la predicción.

```python
df_nominees=pd.read_csv('nominees2021.csv')
X_test=df_nominees.iloc[:,1:]
X_test
```

5. Usando la función `predict_proba` sobre el modelo anterior, predecir las probabilidades que tienen cada película en el `X_test` de ganar el premio Oscar.

```python
win_prob=lr.predict_proba(X_test)
df_win_prob=pd.DataFrame(win_prob,columns=["Probabilidad de Perder","Probabilidad de Ganar"])
df_win_prob.index=df_nominees['Film']
print(df_win_prob)
```

6. ¿Cuál es el nombre de la película con mayor probabilidad de ganar el premio Oscar según su modelo LogisticRegression?

```python
y_predict=lr.predict(X_test)
y_predict

for i in range(len(df_nominees['Film'])):
    if y_predict[i]==1:
        print("La pelicula ganadora es {}".format(df_nominees['Film'][i]))
```

7. Investigue cual fue la película ganadora el año 2021 y comente su predicción. ¿Cuál es la `Accuracy` del modelo?

- Efectivamente la pelicula ganadora del oscar a la mejor pelicula fue Nomadland


```python
from sklearn.metrics import accuracy_score
y_true = [0,0,0,0,1,0,0,0] #Al no tener un CSV con las etiquetas verdaderas creamos un arreglo en base a la investigacion y el modelo programado
accuracy = accuracy_score(y_true,y_predict)
print("Precision del modelo {}".format(accuracy))
```
